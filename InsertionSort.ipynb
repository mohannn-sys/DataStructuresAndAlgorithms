# Insertion Sort
**Algorithm described using Psuedocode**
```
InsertionSort(array A):
    for i from 1 to length(A) - 1 do:
        currentElement = A[i]
        j = i - 1
        
        while j >= 0 and A[j] > currentElement do:
            A[j + 1] = A[j]   // Shift larger elements to the right
            j = j - 1
        
        A[j + 1] = currentElement   // Insert currentElement in its correct position

```
Time Complexity(Worst Case/Average Case) - $O(n^2)$

````python
def insertionSort(arr):
  for i in range(1, len(arr)):
    currentElement = arr[i]
    j = i-1
    while((j >= 0) and (arr[j] > currentElement)):
      arr[j+1] = arr[j]
      j = j-1
    arr[j+1] = currentElement
````

````python
arr = [10,4,8,7,2,6]
insertionSort(arr)
arr

````

````python
import numpy as np
import matplotlib.pyplot as plt
import time
````

````python
arrSizes = [2000, 4000 , 6000, 8000, 10000]
sortTimes = []
for size in arrSizes :
  randomArr = np.random.randint(1, 101, size=size)
  startTime = time.time()
  insertionSort(randomArr)
  endTime = time.time()
  sortTimes.append(endTime - startTime)
````

````python
print(arrSizes)
print(sortTimes)
````

````python
plt.plot(arrSizes, sortTimes, marker='o')
plt.xlabel('Array Sizes')
plt.ylabel('Sorting Time')
plt.title("Sorting time of various array sizes")
plt.show()
````
